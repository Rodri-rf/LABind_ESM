```python
from esm.models.esmc import ESMC
from esm.sdk.api import ESMProtein, LogitsConfig

# Load the 300M model on your GPU
client = ESMC.from_pretrained("esmc_300m").to("cuda")

# Encode a protein sequence
protein = ESMProtein(sequence="MKTAYIAKQRQISFVKSHFSRQLEERLGLIEVQAPILSRVGDGTQDNLSGAEKAVQVKVKALPDAQFEVVHSLAKWKRQTLGQHDFSAGEGLYTHMKALRPDEDRLSPLHSVYVDQWDWERVMGDGERQFSTLKSTVEAIWAGIKATEAAVSEEFGLAPFLPDQIHFVHSQELLSRYPDLDAKGRERAIAKDLGAVFLVGIGGKLSDGHRHDVRAPDYDDWSTPSELGHAGLNGDILVWNPVLEDAFELSSMGIRVDADTLKHQLALTGDEDRLELEWHQALLRGEMPQTIGGGIGQSRLTMLLLQLPHIGQVQAGVWPAAVRESVPSLL")

# Get embeddings
protein_tensor = client.encode(protein)
logits_output = client.logits(
    protein_tensor, 
    LogitsConfig(sequence=True, return_embeddings=True)
)

print(f"Embeddings shape: {logits_output.embeddings.shape}")
print(f"Logits shape: {logits_output.logits.shape}")
```
